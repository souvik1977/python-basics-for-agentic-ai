Step-1: Define the Tools using Pydantic

```python
from pydantic import BaseModel, Field
from typing import Literal

# ---------- Tool 1: Calculator Schema --------
class AddNumbers(BaseModel):
    """Adds two integers together. Use this when the user askes for additon math operations. """
    a: int = Field(description="The first integer number.")
    b: int = Field(description="The second integer number.")

    # ----- Function for Calculator --------------
    #def execute_calculator(a: int, b: int) -> str:
    def execute_calculator(self) -> str:
        return f"Tool Update: The sum of {self.a} and {self.b} is {self.a + self.b}"

# ---------- Tool 2: Weather Schema ------------
class GetWeather(BaseModel):
    """Fetches the current weather metric for a target global city location. """
    city: str = Field(description="The city name, e.g.; 'Kolkata', 'London'")
    unit: Literal["celsius", "farenheit"] = Field(description="Temperature scale",default="celsius")

    # ------ Function for Weather --------------
    #def execute_weather(city: str, unit: str) -> str:
    def execute_weather(self) -> str:
        temp = "32°C" if self.unit == "celsius" else "90°F"
        return f"🌤️ Tool Output: The current weather in {self.city} is {temp} with clear skies."
```

Step-2: Mocking the LLM Payload

```python
# -------------- Simulate when user asks: "What is 452 + 19?"
llm_math_decision = {
    "tool_name": "AddNumbers",
    "arguments": {"a": 452, "b": 19}
}

# --------------- Simulate when user asks: "Check weather in kolkata"

llm_weather_decision = {
    "tool_name": "GetWeather",
    "arguments": {"city": "kolkata", "unit": "celsius"}
}
```

Step-3: The Agentic Tool Routing Engine

```python
from pydantic import ValidationError

def agent_tool_router(llm_payload: dict):
    target_tool = llm_payload.get("tool_name")
    raw_args = llm_payload.get("arguments")

    print(f"🤖 AI Agent requests tool execution: [{target_tool}]")

    try:
        if target_tool == "AddNumbers":
            validate_data = AddNumbers(**raw_args)
            result = validate_data.execute_calculator()
            print(result)
        elif target_tool == "GetWeather":
            validate_data = GetWeather(**raw_args)
            result = validate_data.execute_weather()
            print(result)
            
    
    except ValidationError as ve:
        print(f"Validation Error Occured :: \n{str(ve)}")
```

Step-4: Executing the Logic

```python
print("------------------------ Staring Agentic Function : -------------------")
print("-" * 50)
# 1- Executing the tool_router function with math request
agent_tool_router(llm_math_decision)

print()
print("-" * 50)
# 2- Executing the tool_router function with weather request
agent_tool_router(llm_weather_decision)
```
