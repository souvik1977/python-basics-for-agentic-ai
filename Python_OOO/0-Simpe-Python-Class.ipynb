Example 1 Scenario:
You are writing an automation script to provision AWS S3 or Google Cloud storage buckets.

- Cloud storage buckets must follow strict naming rules, or the cloud provider will reject them.
- Buckets also have a size quota (in Gigabytes) that cannot be negative or exceed enterprise budget limits.

Your Task: Create a CloudBucket class that fulfills these constraints:

1. Properties to create: bucket_name and quota_gb.

2. Validation Rule 1 (bucket_name): The name must be at least 3 characters long. If it is shorter, raise a ValueError with a helpful message.

3. Validation Rule 2 (quota_gb): The storage quota must be a positive integer between 1 and 5000 GB. If it falls outside this range, raise a ValueError.

4. Uniformity: Ensure that when you instantiate the class with CloudBucket("my-data", 500), the validations run immediately during initialization.


```python
class storage_bucket:
    def __init__(self, name, size):
        self.__bucket_name = None
        self.__quota_gb = None

        self.bucket_name = name
        self.quota_gb = size
    
    @property
    def bucket_name(self):
        return self.__bucket_name
    
    @property
    def quota_gb(self):
        return self.__quota_gb
    
    @bucket_name.setter
    def bucket_name(self, name: str):
        name_len = len(name.strip())
        if name_len < 3:
            raise ValueError("Bucket name length is not satisfying the requirement, it must be between 3 - 20 characters long")
        elif name_len > 20:
            raise ValueError("Bucket name length is not satisfying the requirement, it must be between 3 - 20 characters long")

        #Setting up the bucket name
        self.__bucket_name = name

    @quota_gb.setter
    def quota_gb(self, size: int):
        quota_size = size
        if not isinstance(quota_size, int) :
            raise ValueError("Given Bucket Size is not valid. Value must be an integer ranges between 1 - 5000")
        
        if quota_size < 1 and quota_size > 5000:
            raise ValueError("Given Bucket Size is not valid. Value must be an integer ranges between 1 - 5000")

        #Setting up the bucket size
        self.__quota_gb = quota_size
```

```python
# Creating a Storage Bucket and trying to set the name and size
buck_name = input("Enter a bucket name ::").strip()
buck_size = int(input("Enter bucket size [gb] ::").strip())
new_bucket = storage_bucket(buck_name, buck_size)

```

```python
# Display the bucket properties
print(f"Name of the bucket : {new_bucket.bucket_name} and size of the bucket is : {new_bucket.quota_gb} gb")
```

Practice Challenge 2: The API Token Manager

The Scenario: Your automation script connects to server APIs (like Ansible Tower or vCenter) using a secret token.

- To prevent logs from accidentally exposing this sensitive secret token, you want to store it privately.
- When someone prints or views the token property, it should return a masked version (e.g., sk-•••••1234) instead of the raw secret.
- However, you still need a way to change (set) the token when it rotates.

Your Task: Create an ApiConnection class that fulfills these constraints:

Properties to create: 
- environment (public string) and api_token (private gate-keeper property).
- The Setter Rule: When a new token is set, it must strip any accidental whitespaces and ensure it starts with the prefix "sk-". If it doesn't start with "sk-", raise a ValueError.
- The Getter Rule (The Twist): When someone accesses connection.api_token, do not return the raw token. Instead, return a masked string that shows the "sk-" prefix, five dots, and the last 4 characters of the token (e.g., if the token is "sk-secretkey2026", the getter should return "sk-•••••2026").

```python
# Define the class

class appAPI:
    def __init__(self, env: str, api_tok: str):
        self.environment = env 
        self.__api_token = None 

        self.api_token = api_tok
    
    @property
    def api_token(self):
        tok=str(self.__api_token)
        if not tok.startswith("sk-"):
            raise ValueError("API Token Not started with 'sk-'")

        try:
            masked_key = tok[:3] + "•" * (len(tok) - 7) + tok[-4:]
            return masked_key
        except Exception as e:
            raise RuntimeError(f"\n Error Details => Error while masking {str(e)}")

    @api_token.setter
    def api_token(self, new_token: str):
        clean_token = new_token.strip()
        # Check if token is atleast 20 characters long
        if len(clean_token) < 20:
            raise ValueError("\n API Token length is not matching with the requirement (min 20 ch long)")
        # Check if new token is starts with 'sk' 
        
        if not clean_token.startswith("sk-"):
            raise ValueError("\n API Token Not started with 'sk-'")
        
        self.__api_token = clean_token
        
        
```

```python
# Creating the API Class Object
print("Creating a connection of type appAPI")
try:
   
    connection = appAPI('dev', '<Your_API_KEY_Here>')
    print("API Connection initialized successfully!")
    print(f"Environment : {connection.environment}")
    print(f"API Token   : {connection.api_token}")
except Exception as e:
    print(f"Runtime error occured while creating the connection : {str(e)}")
```

```python
# Resetting the key with correct format used
try:
    print("Resetting the API key.....")
    connection.api_token="sk-<Your-New-API-Key-Here>"
except Exception as e:
    print(f"Runtime error occured while Resetting the API key : {str(e)}")
```

```python
# Checking the modified key
print(f"Environment : {connection.environment}")
print(f"API Token   : {connection.api_token}")
```

```python
# Resetting the key with wrong format used
try:
    print("Resetting the API key.....")
    connection.api_token="<Your-Mal-Formed-API-Token-Here-(without sk-)"
except Exception as e:
    print(f"Runtime error occured while Resetting the API key : {str(e)}")
```

```python
# Checking the modified key
print(f"Environment : {connection.environment}")
print(f"API Token   : {connection.api_token}")
```
