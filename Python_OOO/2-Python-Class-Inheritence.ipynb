# Inheritance: Adopting from other class

```python
# Defining Parent Class
class Server:
    def __init__(self, hostname: str, os_ver: str, ip_address: str) -> str:
        self._hostname = hostname
        self._os_ver = os_ver
        self._ip_addr = ip_address
    
    def server_info(self):
        print(f"Hostname = {self._hostname} is Running on {self._os_ver} with IP address of : {self._ip_addr}")
    
    def deploy(self):
        print(f"Deploying on {self._hostname}")
```

```python
# Defining a Linux Server Class which will inherit from Server class

class LinuxServer(Server):
    def __init__(self, hostname: str, os_ver: str, ip_address: str, kernel_ver: str):
        # Call Server initializer
        super().__init__(hostname, os_ver, ip_address)
        self._kernel_ver = kernel_ver
    
    def lin_server_info(self):
        print(f"{self.server_info()} with Kernel Ver : {self._kernel_ver}")
```

```python
# Defining the object for Linux Server
# We want to create 3 Linux Servers

Lin_Servers = []

for i in range(3):
    try:
        print(f"---------------- [{i}] ------------------")
        hostname = input("Hostname : ").strip()
        os_ver = input("OS Version : ").strip()
        ip_addr = input("Server IP : ").strip()
        ker_ver = input("Kernel Version : ").strip()

        if not hostname or not os_ver or not ip_addr or not ker_ver:
            raise ValueError("Hostname, OS-Version, IP Address and Kernel Version are mandatory")

        LinServ_Obj = LinuxServer(hostname, os_ver, ip_addr, ker_ver)
      
        Lin_Servers.append(LinServ_Obj)
    except Exception as e:
        print(f"Error Occured {str(e)}")
    
     

print(f"Successfully Added  {len(Lin_Servers)} Servers")
```

```python
# Displaying the values
for server in Lin_Servers:
    server.lin_server_info()
```
