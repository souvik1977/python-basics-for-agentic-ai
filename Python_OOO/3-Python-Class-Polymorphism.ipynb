# Polymorphism means 'many forms'. In Python OOO concept, different classes can declare functions with same name however with different defintion/body.

Exact same name [but] completely different execution logic

```python
# Importing 'ipaddress' class for IP Validation
import ipaddress
```

```python
# Defining a Server class

# we will use @property (Getter) and @setter gateways to set the property and read it back

class Server:
    def __init__(self, hostname, ip_addr):
        self.__hostname = hostname
        self.__ip_addr = None
        self.ip_addr = ip_addr

    @property
    def hostname(self):
        return self.__hostname
    
    @property
    def ip_addr(self):
        return self.__ip_addr
    
    @ip_addr.setter
    def ip_addr(self, new_ip):
        try:
            clean_ip = new_ip.strip()
            ipaddress.ip_address(clean_ip)
        except ValueError as ve:
            raise ValueError(f"Invalid IP Format Given for {new_ip}")
        
        self.__ip_addr = clean_ip
```

```python
# Defining a LinuxServer Class which will inherit from Server class
class LinuxServer(Server):
    def __init__(self, hostname, ip_addr):
        super().__init__(hostname, ip_addr)
    
    def get_info(self):
        print(f"Linux Server Hostname : {self.hostname} IP Address : {self.ip_addr}")

```

```python
# Defining a Windows class which will inherit from Server class
class WindowsServer(Server):
    def __init__(self, hostname, ip_addr):
        super().__init__(hostname, ip_addr)
    
    def get_info(self):
        print(f"Windows Server Hostname : {self.hostname} IP Address : {self.ip_addr}")
```

```python
# Defining Linux Server Object
LS1 = LinuxServer("webserver1", "192.168.50.50")

```

```python
# Defining Windows SErver Object
WS1 = WindowsServer("mysqldb1", "192.168.40.25")
```

```python
# Displaying Details of our Linux Server using 'get_info()' function
LS1.get_info()

# Displaying Details of our Linux Server using 'get_info()' function
WS1.get_info()
```

```python
# Changing the IP Address of Linux_Server
print("Changing Web Server IP to separate VLan/Subnet\n")
new_web_ip = input("Enter a new IP address :").strip()
LS1.ip_addr = new_web_ip
```

```python
# Displaying Details of our Linux Server using 'get_info()' function
print("\nPost Getting the New IP, here are the details of Linux Server\n")
LS1.get_info()
```

Above We tried to access Privage Properties outside of the Class using Getter [@property] and Setter [@setter]. Now we will implement java/classic style implementation by using functions

```python
class Server:
    def __init__(self, hostname, ip_addr):
        self.__hostname = hostname
        self.__ip_addr = None
        
        # We must call the set_ip_addr() explicitly
        self.set_ip_addr(ip_addr)


    def get_hostname(self):
        return self.__hostname
    

    def get_ipaddress(self):
        return self.__ip_addr
    
    
    def set_ip_addr(self, new_ip):
        try:
            clean_ip = new_ip.strip()
            ipaddress.ip_address(clean_ip)
        except ValueError as ve:
            raise ValueError(f"Invalid IP Format Given for {new_ip}")
        
        self.__ip_addr = clean_ip
```

```python
# Defining a LinuxServer Class which will inherit from Server class
class LinuxServer(Server):
    def __init__(self, hostname, ip_addr):
        super().__init__(hostname, ip_addr)
    
    def get_info(self):
        print(f"Linux Server Hostname : {self.get_hostname()} IP Address : {self.get_ipaddress()}")
```

```python
# Defining a Windows class which will inherit from Server class
class WindowsServer(Server):
    def __init__(self, hostname, ip_addr):
        super().__init__(hostname, ip_addr)
    
    def get_info(self):
        print(f"Windows Server Hostname : {self.get_hostname()} IP Address : {self.get_ipaddress()}")
```

```python
# Defining Linux Server Object
LS1 = LinuxServer("webserver1", "192.168.20.50")

# Defining Windows SErver Object
WS1 = WindowsServer("mysqldb1", "192.168.6.25")
```

```python
# Displaying Linux and Windows Servers Properties
# Displaying Details of our Linux Server using 'get_info()' function
LS1.get_info()

# Displaying Details of our Linux Server using 'get_info()' function
WS1.get_info()
```

```python
# Setting new IP address for Windows Server

print("Changing Server IP to separate VLan/Subnet\n")
new_ip = input("Enter a new IP address :").strip()
WS1.set_ip_addr(new_ip)

# Displaying Details of our Windows Server using 'get_info()' function
# Displaying Details of our Linux Server using 'get_info()' function
print("\nPost Getting the New IP, here are the details of Server\n")
WS1.get_info()
```
