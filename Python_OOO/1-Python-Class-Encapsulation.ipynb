# Encapsulation : Data is private and not accessible outside of the class
# Private data are not accessible outside of the class, this needs an class object

# In python we use "__" (double under score) to make a variable private and protected

# we will use @property (Getter) and @setter gateways to set the property and read it back in case we want to access private class properties directly, from outside of a class.
# If we will use class functions to access private properties then @property and @setter are not required 

```python
import ipaddress

class LinuxServer:
    def __init__(self, hostname: str, ipaddr: str):
        self.__hostname = hostname
        self.__ip_addr = None
        self.ip_addr = ipaddr

    # The @getter to read the ip address safely
    @property
    def ip_addr(self):
        return self.__ip_addr
    
    def hostname(self):
        return self.__hostname
    
    # The @setter Validate the data before changing the private variable
    @ip_addr.setter
    def ip_addr(self, new_ip):
        try:
            cleaned_ip = str(new_ip).strip()
            # Uses Python's built-in ipaddress module to validate format
            ipaddress.ip_address(cleaned_ip)
        except ValueError:
            raise ValueError(f"Critical : '{new_ip}' is in invalid format")
        
        self.__ip_addr = cleaned_ip

# Defining the Object
server = LinuxServer("web-server-1", "192.168.1.10")

print(f"Server Details : \n\tHostname : {server.hostname()}\n\tIP : {server.ip_addr}")
```

```python
# Trying to override the ipaddress
server.__ip_addr = "broken"

print(f"Details Post wrong IP Address format \nServer Details : \n\tHostname : {server.hostname()}\n\tIP : {server.ip_addr}")
```

```python
# Trying to override the ipaddress
server.ip_addr = "192.168.9.25"

print(f"Details Post Overriding the IP successfully :: \nServer Details : \n\tHostname : {server.hostname()}\n\tIP : {server.ip_addr}")
```
